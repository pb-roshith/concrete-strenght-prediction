```python
import numpy as np
import pandas as pd
import keras
```

```python
concrete_data = pd.read_csv('https://s3-api.us-geo.objectstorage.softlayer.net/cf-courses-data/CognitiveClass/DL0101EN/labs/data/concrete_data.csv')
concrete_data.head()
```

```python
concrete_data.shape
```

```python
concrete_data.describe()
```

```python
concrete_data.isnull().sum()
```

```python
concrete_data.columns
```

```python
columns = concrete_data.columns
predictor = concrete_data[columns[ columns != 'Strength']]
target = concrete_data['Strength']
```

```python
predictor.head()
```

```python
target.head()
```

```python
predictor_norm = (predictor - predictor.mean()) / predictor.std()
predictor_norm.head()
```

```python
predictor_norm.shape[1]
```

```python
n_cols = predictor_norm.shape[1]
```

```python
from keras.models import Sequential
from keras.layers import Dense
```

```python
def regression_model():
  model = Sequential()
  model.add(Dense(50, activation='relu', input_shape=(n_cols,)))
  model.add(Dense(50, activation='relu'))
  model.add(Dense(1))
  model.compile(optimizer='adam', loss='mean_squared_error')
  return model
```

```python
model = regression_model()
model.fit(predictor_norm, target, validation_split=0.3, epochs=100, verbose=2)
```

```python
model.predict([[2.476712,	-0.856472,	-0.846733,	-0.916319,	-0.620147,	0.862735,	-1.217079,	-0.279597]])
```
